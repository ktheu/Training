### Rekursion

Alle Funktionen sollen rekursiv implementiert werden.

```python
def fakultaet(n):
    '''
    n: int
    returns: n!
    
    >>> fakultaet(0)
    1

    >>> fakultaet(6)
    720

    >>> fakultaet(13)
    6227020800
    '''
    pass

import doctest
doctest.run_docstring_examples(fakultaet,globals(),optionflags=doctest.NORMALIZE_WHITESPACE) 
```
